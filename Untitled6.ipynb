```python
import matplotlib.pyplot as plt
import numpy as np

# Paramètres du signal
t = np.linspace(0, 1, 500)
carrier_freq = 100
message_freq = 2
carrier_signal = np.sin(2 * np.pi * carrier_freq * t)
message_signal = np.sin(2 * np.pi * message_freq * t)

# Création de la figure et des sous-graphiques
fig, axs = plt.subplots(6, 1, figsize=(10, 14))

# Signal porteur et message combinés
axs[0].plot(t, carrier_signal, label='Carrier Signal', color='blue')
axs[0].plot(t, message_signal, label='Message Signal', color='orange', linestyle='--')
axs[0].set_title('Carrier and Message Signal')
axs[0].legend()

# Modulation d'amplitude d'impulsion (PAM)
pam_signal = carrier_signal * message_signal
axs[1].plot(t, pam_signal, label='PAM Signal', color='green')
axs[1].set_title('Pulse Amplitude Modulation (PAM)')
axs[1].legend()

# Modulation de largeur d'impulsion (PWM)
pwm_signal = np.where(message_signal > 0, 1, -1)
axs[2].plot(t, pwm_signal, label='PWM Signal', color='red')
axs[2].set_title('Pulse Width Modulation (PWM)')
axs[2].legend()

# Modulation de position d'impulsion (PPM)
ppm_signal = np.zeros_like(t)
for i in range(len(t)):
    if i % 50 == 0:
        ppm_signal[i:i+10] = 1
axs[3].plot(t, ppm_signal, label='PPM Signal', color='purple')
axs[3].set_title('Pulse Position Modulation (PPM)')
axs[3].legend()

# Modulation de fréquence d'impulsion (PFM)
pfm_signal = np.zeros_like(t)
frequency_variation = np.sin(2 * np.pi * message_freq * t)  # Variation de fréquence
for i in range(len(t)):
    if i % int(10 + 5 * frequency_variation[i]) == 0:  # Fréquence variable
        pfm_signal[i:i+5] = 1
axs[4].plot(t, pfm_signal, label='PFM Signal', color='brown')
axs[4].set_title('Pulse Frequency Modulation (PFM)')
axs[4].legend()

# Ajustement de l'espacement entre les sous-graphiques
plt.tight_layout()

# Affichage de la figure
plt.show()
```
